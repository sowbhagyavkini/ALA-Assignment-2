```python
import numpy as np
import pandas as pd

# Given dataset
data = pd.DataFrame({
    'HR': [76, 74, 72, 78],
    'BP': [126, 120, 118, 136],
    'Temp': [38.0, 38.0, 37.5, 37.0]},
    index=['patient-1','patient-2','patient-3','patient-4']
)

data.head()
```

a.which patient is farthest from the rest?

to find the distance between two points we can use euclidean distance formula


```python
def euclidean_distance(p1, p2):
    return np.sqrt(np.sum((p2 - p1) ** 2))

total_distance = 0
farthest_patient = 0
maximum_distance = 0

for p1, data1 in data.iterrows(): # take data from patient1 . iterrow generates an iterator object of dataframe
    total_distance = 0
    for p2, data2 in data.iterrows(): #take data from patient2.
            distance = euclidean_distance(data1, data2) #calculates the Euclidean distance between two data points data1 and data2.
            total_distance += distance
    if total_distance > maximum_distance:
        maximum_distance = total_distance
        farthest_patient = p1 #calculate distance between each patient

print("Farthest patient from the rest:", farthest_patient)
```

which two patients are near?

```python
minimum_distance = np.inf #inf represents positive and negative infinite
near_patients =0


for p1, data1 in data.iterrows():
    for p2, data2 in data.iterrows():
        if p1 != p2:
            distance = euclidean_distance(data1, data2)
            if distance < minimum_distance:
                minimum_distance = distance
                near_patients = (p1, p2)

print("Near patients:", near_patients)
```
